For loading a csv file

```python
import pandas as pd
```

```python
df = pd.read_csv("supermarkets.csv")
```

```python
df
```

For accessing specific columns of file

```python
df['Chain']
```

```python
df.Chain
```

If your data doesn't has headers then do this

```python
df.shape
```

```python
df2 = pd.read_json("095 supermarkets.json")
```

---------------------

For accessing xlsx files

```python
df3 = pd.read_excel("supermarkets.xlsx")
```

```python
df3
```

```python
df3.set_index('ID')
```

To give sheetname do this 
pd.read_excel("supermarkets.xlsx", sheetname=0)
